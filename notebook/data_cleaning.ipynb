```python
import pandas as pd
```

```python
def data_cleaning(df: pd.DataFrame, hashtable: dir) -> pd.DataFrame:
    # remove the uncessary columns
    df = df.drop(['ID'], axis=1)
    
    # deal with the missing values, mark as -1 (we treat unknown as same as other)
    df['EDUCATION'].replace({0:-1,4:-1,5:-1,6:-1}, inplace=True)
    
    # 1: school, 2: high school, 3: university
    df['EDUCATION'].replace({1:1,2:3,3:2}, inplace=True)
    
    # deal with the missing values, mark as -1 (we treat unknown as same as other)
    df['MARRIAGE'].replace({0:-1,3:-1}, inplace=True)
    
    # update the name of the cleaned data according the hashtable
    df.rename(columns=hashtable, inplace=True)
    
    # save the cleaned data
    df.to_csv('../data/cleaned_data.csv', index=False)
    return df
```

```python
hashtable = {
        'PAY_0': 'Repayment_Status_Sept',
        'PAY_2': 'Repayment_Status_Aug',
        'PAY_3': 'Repayment_Status_Jul',
        'PAY_4': 'Repayment_Status_Jun',
        'PAY_5': 'Repayment_Status_May',
        'PAY_6': 'Repayment_Status_Apr',
        'BILL_AMT1': 'Bill_Amount_Sept',
        'BILL_AMT2': 'Bill_Amount_Aug',
        'BILL_AMT3': 'Bill_Amount_Jul',
        'BILL_AMT4': 'Bill_Amount_Jun',
        'BILL_AMT5': 'Bill_Amount_May',
        'BILL_AMT6': 'Bill_Amount_Apr',
        'PAY_AMT1': 'Pay_Amount_Sept',
        'PAY_AMT2': 'Pay_Amount_Aug',
        'PAY_AMT3': 'Pay_Amount_Jul',
        'PAY_AMT4': 'Pay_Amount_Jun',
        'PAY_AMT5': 'Pay_Amount_May',
        'PAY_AMT6': 'Pay_Amount_Apr',
        'default payment next month': 'Default_Payment'
    }

df = pd.read_csv('../data/default_of_credit_card_client.csv')
df = data_cleaning(df, hashtable)
df.head()
```
