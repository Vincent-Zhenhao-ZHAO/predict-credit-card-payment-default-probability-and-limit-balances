```python
import pandas as pd
import sys
```

```python
def get_data(PATH: str, file_name: str) -> pd.DataFrame:
    try:
        df = pd.read_excel(PATH+file_name, header=1)
        # fill empty space with _
        csv_filename = file_name.replace(" ","_")
        df.to_csv(PATH+csv_filename[:-5]+".csv", index=False)
        df = pd.read_csv(PATH+csv_filename[:-5]+".csv")
        return df
    except:
        print("Error: file not found")
        sys.exit(1)

df = get_data("../data/", "default of credit card clients.xls")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd  # Assuming you will load the data into a pandas DataFrame

# Assuming 'df' is your DataFrame and it contains the necessary columns
# Replace 'df' with the actual name of your DataFrame variable

skewness_dict = {
    'PAY': df[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']].skew().values,
    'BILL_AMT': df[['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']].skew().values,
    'PAY_AMT': df[['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']].skew().values
}

# Create a line chart for each feature set
plt.figure(figsize=(14, 6))

# X-axis values
time_periods = [1, 2, 3, 4, 5, 6]

# Plot each line
for feature, skewness in skewness_dict.items():
    plt.plot(time_periods, skewness, marker='o', label=f'Skewness of {feature}')

plt.legend()
plt.title('Skewness of PAY, BILL_AMT, and PAY_AMT Variables Over Time')
plt.xlabel('Time Period')
plt.ylabel('Skewness')
plt.xticks(time_periods)  # Ensure x-axis ticks show each time period
plt.grid(True)  # Optional: Add grid for better readability

# Show plot
plt.show()
```

```python
sns.set_theme(style="white")
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
